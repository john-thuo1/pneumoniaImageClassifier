```python
# Load Data
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator
```

```python
data = tf.keras.utils.image_dataset_from_directory('data',color_mode='grayscale',image_size=(224, 224),)

data_iterator = data.as_numpy_iterator()
```

```python
batch = data_iterator.next()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    if batch[1][idx] == 0:
        ax[idx].title.set_text('Normal')
    else:
        ax[idx].title.set_text('Pneumonia')
```

```python
# Scale the data 
# We standardize the pixels to range 0-1

data = data.map(lambda x,y: (x/255, y))
data.as_numpy_iterator().next()
```

```python
# Split the data 
train_size = int(len(data)*.7)
# Used During Training
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

## Building the Network


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
from sklearn.metrics import classification_report,confusion_matrix
from keras.callbacks import ReduceLROnPlateau
```

```python
# keras sequential model
model = Sequential()
model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (224,224,1)))
model.add(MaxPooling2D())

model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(MaxPooling2D())

model.add(Flatten())
model.add(Dense(units = 64 , activation = 'relu'))
model.add(Dense(units = 1 , activation = 'sigmoid'))
```

# Compile network

```python
# Using binary_crossentropy because this is a binary classification problem with 2 classes
'''
Optimizers: we used the Adam optimizer to update our weights during training. 
When training your network, you should experiment with other optimizers and their learning rate.
'''

# model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Info about our layers 
# Trainable parameters are the ones that will be updated during training to optimize the model, 
# while non-trainable parameters are fixed and typically used as biases or normalization parameters.
# Build the model
model.summary()
```

# Training from Scratch


```python
# Use model.fit()
# Specify batch_size, epochs, and validation data
```

```python
# Using a batch_size of 32, 50 epochs 
# Fit the model with augmented data

trainedM = model.fit(train,batch_size=32,epochs=50,validation_data=val, callbacks=[tensorboard_callback])
```

# Transfer Learning using VGG-16

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Lambda
from tensorflow.keras.layers import Lambda
from tensorflow.keras import backend as K

# Load the VGG16 model without the top layers (include_top=False)
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the VGG16 model layers weights
# Don't train the VGG16 layers as they are already trained.
for layer in vgg_model.layers:
    layer.trainable = False

# Create a new model using the VGG16 model as a base
Tmodel = Sequential()
Tmodel.add(Lambda(lambda x: K.repeat_elements(x, rep=3, axis=-1), input_shape=(224, 224, 1)))
Tmodel.add(vgg_model)




# Adding  a new classifier layer on top of the base model
Tmodel.add(Flatten())
Tmodel.add(Dense(units=64, activation='relu'))
Tmodel.add(Dense(1, activation='sigmoid'))

# Compile the model with the desired optimizer and loss function
Tmodel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model using the data
hist = Tmodel.fit(train, epochs=50, batch_size=32, validation_data=val)
```
